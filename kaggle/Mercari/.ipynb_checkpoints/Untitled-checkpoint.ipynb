```python
train['brand_name']=train.brand_name.fillna('nan')
test['brand_name']=test.brand_name.fillna('nan')
```

```python
train['category_name']=train.category_name.fillna('nan')
test['category_name']=test.category_name.fillna('nan')
```

```python
train['cate_list']=train.category_name.apply(lambda x:x.split('/'))
test['cate_list']=test.category_name.apply(lambda x:x.split('/'))
```

```python
train['cate_num']=train.cate_list.apply(lambda x:len(x))
test['cate_num']=test.cate_list.apply(lambda x:len(x))
```

```python
train['cate1']=train.cate_list.apply(lambda x:x[0])
train['cate2']=train.cate_list.apply(lambda x:x[1] if len(x)>1 else 'nan' )
```
